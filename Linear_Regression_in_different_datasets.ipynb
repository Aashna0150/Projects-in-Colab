```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/tvmarketing.csv")
df
```

```python
x=df.iloc[:,0:1].values
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#when tv value is 50 what would be the new sales value?
new=lr.predict([[50]])
new
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred)*100)
```

```python
import numpy as np
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
import pandas as pd
df=pd.read_csv("/content/dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.dtypes
```

```python
df.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
df.drop(["Name"],axis=1,inplace=True)
```

```python
df['Location'].value_counts()
```

```python
df["Fuel_Type"].value_counts()
```

```python
df["Transmission"].value_counts()
```

```python
df["Owner_Type"].value_counts()
```

```python
dumm=pd.get_dummies(df[["Location","Fuel_Type","Transmission","Owner_Type"]])
dumm
```

```python
dumm.columns
```

```python
#from sklearn.preprocessing import LabelEncoder
#lr=LabelEncoder()
#dumm["Location_Coimbatore"]=lr.fit_transform(dumm["Location_Ahmedabad"])
#can use label encoder for all the columnsin the same way
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in dumm.columns:
  dumm[i]=le.fit_transform(dumm[i])
```

```python
dumm
```

```python
df
```

```python
#we drop these columns because we got the dummies of corresponding columns
df.drop(["Location","Fuel_Type","Transmission","Owner_Type"],axis=1,inplace=True)
```

```python
df
```

```python
#to combine df and dumm dataframes
df1=pd.concat([df,dumm],axis=1)
df1
```

```python
df1.isna().sum()
```

```python
df1["Mileage"]=df1["Mileage"].str.replace("km/kg","")
```

```python
df1["Mileage"]=df1["Mileage"].str.replace("kmpl","")
```

```python
df1["Engine"]=df1["Engine"].str.replace("CC","")
```

```python
df1["Power"]=df1["Power"].str.replace("bhp","")
```

```python
df1["Power"]=df1["Power"].str.replace("null","0")
```

```python
df1
```

```python
df1.info()
```

```python
#since only str.replace values exist mileage,power etc values cannot be converted to flot or int directly so we use .astype(flot/int)
df1["Mileage"]=df1["Mileage"].astype(float)
df1["Engine"]=df1["Engine"].astype(float)
df1["Power"]=df1["Power"].astype(float)
```

```python
df1
```

```python
df1.info()
```

```python
import numpy as np
df1.loc[df1.Power==0,"Power"]=np.nan
```

```python
df1.isnull().sum()
```

```python
df1["Mileage"].value_counts()
df1["Engine"].value_counts()
df1["Power"].value_counts()
df1["Seats"].value_counts()
```

```python
df1["Mileage"]=df1["Mileage"].fillna(df1["Mileage"].mean())
df1["Engine"]=df1["Engine"].fillna(df1["Engine"].mean())
df1["Power"]=df1["Power"].fillna(df1["Power"].mean())
df1["Seats"]=df1["Seats"].fillna(df1["Seats"].median())
```

```python
df1
```

```python
y=df1["Price"].values
y
```

```python
x=df1.drop(["Price"],axis=1).values
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
print(r2_score(y_test,y_pred)*100)
```

```python
print(lr.coef_)
print(lr.intercept_)
```
